```python
cbb22 = pd.read_csv("cbb22.csv")
```

```python
cbb22.plot(kind = 'scatter', x = 'ADJOE', y = 'BARTHAG')

plt.show()
```

```python
# Plot first y axis
plt.plot(cbb22['ADJOE'], cbb22['BARTHAG'], label='BARTHAG', marker='o', linestyle='None')

plt.twinx()  # Create a secondary y-axis
plt.plot(cbb22['ADJOE'], cbb22['W'], label='Wins', marker='x', color='red', linestyle='None')

plt.xlabel('ADJOE')
plt.ylabel('Wins')
plt.ylabel('BARTHAG')
plt.legend(loc='upper left')

plt.show()
```
